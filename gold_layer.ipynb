```python
# CELLA 4 - Analisi dati
spark.sql(f"""
    SELECT
        CASE 
            WHEN consumo_misto_wltp_l100km IS NULL AND consumo_elettrico_kwh100km IS NOT NULL THEN 'Elettrico'
            WHEN autonomia_elettrica_km IS NOT NULL THEN 'PHEV'
            WHEN motorizzazione LIKE '%Hybrid%' THEN 'Ibrido'
            ELSE 'Termico'
        END AS tipo_alimentazione,
        COUNT(*)                                AS n_veicoli,
        ROUND(AVG(emissioni_co2_gkm), 1)        AS media_co2_gkm,
        ROUND(AVG(accelerazione_0_100_s), 2)    AS media_0_100_s,
        ROUND(AVG(autonomia_km), 0)             AS media_autonomia_km
    FROM delta.`{DELTA_PATH}`
    GROUP BY 1
    ORDER BY media_co2_gkm
""").display()
```
